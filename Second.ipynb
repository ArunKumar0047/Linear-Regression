```python
import pandas as pd
import numpy as np
import random
import matplotlib.pyplot as plt
```

```python
np.random.seed(47)
x=np.random.randn(1000)
g=5
y=(g*np.random.randn(1000))+x
```

```python
def gradient_descend(LR,iterations,x,y):
    m=5
    c=0
    n = x.shape[0]
    for i in range(iterations): 
        Y = m*x + c #predicting Line of best fit
        D_m = (-2/n) * np.sum(x * (y - Y)) #Partial derivative of Least square w.r.t slope
        D_c = (-2/n) * np.sum(y - Y) #Partial derivative of Least square w.r.t intercept
        m = m - (LR * D_m  )
        c = c - (LR * D_c) 
    return m,c
```

```python
m,c=gradient_descend(0.01,1000,x,y)
Y_pred=m*x + c
plt.scatter(x,y,s=7,c='b')
plt.plot(x,Y_pred)
```
